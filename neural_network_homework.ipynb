```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
epochs = 1000
learning_rate = 0.1
```

```python
x = tf.reshape(tf.constant([0.05, 0.1]), (1, 2))
x = tf.concat([x, tf.ones((x.shape[0], 1))], axis=1)

x
```

```python
y = tf.constant(0.5)
```

```python
tf.random.set_seed(1)
w1 = tf.Variable(
    tf.random.uniform((3, 2), minval=0, maxval=1)
)
w1
```

```python
w2 = tf.Variable(
    tf.random.uniform((3, 1), minval=0, maxval=1)
)
w2
```

```python
for i in range(epochs):
    with tf.GradientTape() as tape:
        h = tf.sigmoid(tf.matmul(x, w1))
        h = tf.concat([h, tf.ones((h.shape[0], 1))], axis=1)
        y_pred = tf.sigmoid(tf.matmul(h, w2))
        MSE = (y - y_pred)**2
        
    dMSE = tape.gradient(MSE, [w1, w2])

    w1.assign_sub(learning_rate * dMSE[0]) 
    w2.assign_sub(learning_rate * dMSE[1])
    
```
